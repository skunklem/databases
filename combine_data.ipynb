```python
from pathlib import Path
import pandas as pd
from Bio import SeqIO
```

```python
plate="03-11-22-1-A4.1"
gisaid_csv = Path(f"{plate}-All-Gisaid.csv")
fasta = Path(f"{plate}-All-Gisaid.fasta")
main_csv = Path(f"Sequencing-report-{plate}-All.csv")
accession_file = Path("03-11-22-1-A4.1-gisaid_uploader.log")
```

```python
g = pd.read_csv(gisaid_csv)
g
```

```python
def getFasta(seqID,fastas):
    seq = fastas.get(seqID)
    if not seq: raise Exception(f"MissingSequence for {seqID}")
    return seq
```

```python
samples = {record.id:record.seq for record in SeqIO.parse(fasta,"fasta")}
samples
g["consensus"] = g["covv_virus_name"].apply(lambda x: getFasta(x,samples))
g
# pd.from_dict(samples)
```

```python
m = pd.read_csv(main_csv)
m
```

```python
g["Sample #"] = g["covv_virus_name"].apply(lambda x: x.split("/")[2].replace("NC-",""))
g
```

```python
a = pd.read_csv(accession_file,sep=";",header=None,names=["covv_virus_name","accession"]).dropna()
a['accession'] = a['accession'].apply(lambda x: x.strip())
g = g.merge(a,on="covv_virus_name")
g
```

```python
merged = m.merge(g, on="Sample #").merge(a)
merged.columns
```

```python
m['Sequence date']
```

```python
merged
```

```python
df = merged.rename(columns={
    "Sample #":"sample_id",
#     "Source Lab":"subm_lab_id",
    "Sequence date":"covv_location",
    "covv_location":"county",
    "Reads on Barcode":"read_count",
    "Pango lineage":"pango_lineage",
    "Pangolin QC":"pango_qc",
    "Test date":"test_date"
})
df
```

```python
lab = pd.DataFrame.from_dict({
    "lab_id":[1,2,3,4],
    "name":["UNC Charlotte COVID-19 Testing Lab","UNC Charlotte Environmental Monitoring Laboratory","Mission Hospital, LLLP","StarMed Healthcare"],
    "address":["9201 University City Blvd, Charlotte NC 28223","9201 University City Blvd, Charlotte NC 28223","509 Biltmore Avenue, B201.01, Asheville, NC 28801-4601","1891 Scott Futrell Drive, Charlotte NC 28208"]
})
lab = lab
lab
```

```python
df["subm_lab_id"] = df.merge(lab[["lab_id","name"]],
                             left_on="covv_subm_lab", right_on="name")[["lab_id"]]
df["orig_lab_id"] = df.merge(lab[["lab_id","name"]],
                             left_on="covv_orig_lab", right_on="name")[["lab_id"]]
df["run_id"] = 54
df["host_id"] = 1
df
```

```python
host = pd.DataFrame.from_dict({"host_id":[1],"species":["Human"]})
host
```

```python
run_info =  pd.DataFrame.from_dict({"run_id":[54],"protocol_id":["11"],"plate_id":[plate],"seq_date":["2022-03-11"]})
run_info
```

```python
protocol = pd.DataFrame.from_dict({"protocol_id":["11"],
                                   "sequencing_tech":["Oxford Nanopore PromethION"],
                                   "assembly_method":["ARTIC Medaka pipeline"]})
protocol
```

```python
gisaid = df[["accession","sample_id","covv_virus_name"]]
gisaid
```

```python
sample = df[["sample_id","subm_lab_id","orig_lab_id","run_id","read_count","test_date","pango_lineage","pango_qc","consensus","county","host_id"]]
sample
```

```python
sample.to_csv(f"sample_{plate}.csv",index=False)
lab.to_csv(f"lab_{plate}.csv",index=False)
run_info.to_csv(f"run_info_{plate}.csv",index=False)
gisaid.to_csv(f"gisaid_{plate}.csv",index=False)
host.to_csv(f"host_{plate}.csv",index=False)
protocol.to_csv(f"protocol_{plate}.csv",index=False)
```
